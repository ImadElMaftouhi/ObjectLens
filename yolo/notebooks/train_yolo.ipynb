```python
import os
os._exit(0)
```

```python
from ultralytics import YOLO

data_path = "/kaggle/working/data_kaggle.yaml"

model = YOLO("yolov8n.pt")

model.train(
    data=data_path,
    epochs=100,
    imgsz=640,
    batch=64,                 # or 64 if it fits
    workers=4,
    project="runs_imagenet15",
    name="yolov8n_imagenet15",
    pretrained=True,
    patience=10,
    verbose=True,
    device="0,1"              # use both T4 GPUs
)
```
